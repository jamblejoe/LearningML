## Algorithms

```julia
# gradient descent
function gradient_descent_step(dL::Function, theta::Real, eta::Real)
    theta_next = theta - eta * dL(theta)
    return theta_next
end

function gradient_descent(dL::Function, start; eta::Real=1., max_steps::Integer=100)
    
    x = start
    xs = [start]
    
    step_counter = 1
    while step_counter < max_steps
        x = gradient_descent_step(dL, x, eta)
        push!(xs, x)
        step_counter += 1
    end
    
    return xs
end

# momentum gradient descent
function momentum_gradient_descent_step(dL::Function, theta::Real, m::Real, eta::Real)
    # m_i
    m_next = g * m- (1-g) * eta * dL(theta)
    # theta_{i+1}
    theta_next = theta + m_next
    
    return (theta_next, m_next)
end

function momentum_gradient_descent(dL::Function, start; eta::Real=1., max_steps::Integer=100)
    
    x = start
    m = 0
    xs = [start]
    ms = []
    
    step_counter = 1
    while step_counter < max_steps
        x, m = momentum_gradient_descent_step(dL, x, m, eta)
        
        push!(xs, x)
        push!(ms, m)
        
        step_counter += 1
    end
    
    return xs, ms
end

# (super) acceleration
function accelerated_gradient_descent_step(dL::Function, theta::Real, m::Real, eta::Real, nu::Real)
    # m = m_{i-1}
    # theta = theta_i
    
    # m_i
    m_next = g * m- (1-g) * eta * dL(theta + nu * m)
    # theta_{i+1}
    theta_next = theta + m_next
    
    return (theta_next, m_next)
end

function accelerated_gradient_descent(dL::Function, start; eta::Real=1., nu::Real=1., max_steps::Integer=100)
    
    x = start
    m = 0
    xs = [start]
    ms = []
    
    step_counter = 1
    while step_counter < max_steps
        x, m = accelerated_gradient_descent_step(dL, x, m, eta, nu)
        
        push!(xs, x)
        push!(ms, m)
        
        step_counter += 1
    end
    
    return xs, ms
end
```

```julia
# functions
L = x -> 1/2 * x^2
dL = x -> x
```

```julia
# parameters
# starting position
x_1 = 1.
const g = 0.9
```

```julia
let 
    eta = 0.5 
    xs = gradient_descent(dL, x_1;eta=eta)
    
    total_steps = length(xs)
    final_x = xs[end]
    final_L_x = L(xs[end])
    
    x_min = minimum(xs) - 0.1
    x_max = maximum(xs) + 0.1
    x_values = range(x_min;stop=x_max,length=20)
    plt.plot(x_values, L.(x_values))
    plt.plot(xs, L.(xs); linestyle="--", marker="o", markersize=4., color="red")
    
    
    sleep(0.1)
    @show total_steps
    @show final_x
    @show final_L_x
    
    nothing
end
```

```julia
let 
    eta = 0.5
    xs,_ = momentum_gradient_descent(dL, x_1;eta=eta)
    
    total_steps = length(xs)
    final_x = xs[end]
    final_L_x = L(xs[end])
    
    x_min = minimum(xs) - 0.1
    x_max = maximum(xs) + 0.1
    x_values = range(x_min;stop=x_max,length=20)
    plt.plot(x_values, L.(x_values))
    plt.plot(xs, L.(xs); linestyle="--", marker="o", markersize=4., color="red")
    
    
    sleep(0.1)
    @show total_steps
    @show final_x
    @show final_L_x
    
    nothing
end
```

```julia
let 
    eta = 0.5
    nu = 0.5
    xs,_ = accelerated_gradient_descent(dL, x_1;eta=eta, nu=nu)
    
    total_steps = length(xs)
    final_x = xs[end]
    final_L_x = L(xs[end])
    
    x_min = minimum(xs) - 0.1
    x_max = maximum(xs) + 0.1
    x_values = range(x_min;stop=x_max,length=20)
    plt.plot(x_values, L.(x_values))
    plt.plot(xs, L.(xs); linestyle="--", marker="o", markersize=4., color="red")
    
    
    sleep(0.1)
    @show total_steps
    @show final_x
    @show final_L_x
    
    nothing
end
```
